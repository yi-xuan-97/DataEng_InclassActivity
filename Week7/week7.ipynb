```python
options = ['Loudoun County','Washington County','Harlan County','Malheur County']
options1 = ['Virginia','Oregon','Kentucky']
temp = census[census['County'].isin(options) & census['State'].isin(options1)]
data = temp.groupby(['County']).mean()
county_info = data[['TotalPop','Poverty','IncomePerCap']]
county_info.head
```

```python
covid['date'] = pd.to_datetime(covid['date'])
covid['day_of_week'] = covid['date'].dt.day_name()
covid['month'] = pd.DatetimeIndex(covid['date']).month
covid['year'] = pd.DatetimeIndex(covid['date']).year
covid.head(200)
```

```python
options = ['Malheur']
options1 = ['Oregon']
options2 = [1,2,8]

temp = covid[covid['county'].isin(options) & covid['state'].isin(options1)]

data = temp.groupby(['county','month','year']).sum()
data[['cases','deaths']].head(200)
# idx = pd.IndexSlice
# A = data.loc[:,idx[:,'Melheur']]
# B = data.loc[:,idx[:,'2021']]
# A.head()
```

```python
data = covid[['state','county','fips']]
temp = data.drop_duplicates()
temp.head(100)

data = covid.groupby(['county','month','year','county','state']).sum()
COVID_monthly = data[['cases','deaths']]
COVID_summary = pd.concat([temp, COVID_monthly],axis=1)
COVID_summary.head()
```

```python
def line_prepender(filename,line):
    with open(filename, 'r+') as f:
        content = f.read()
        if content.startswith(","):
            f.seek(0, 0)
            f.write(line.rstrip('\r') + content)
line_prepender('COVID_county_data.csv',"INDEX")
```
